#### Various kind of distances in machine learning
In machine learning and data analysis, various distance metrics are used depending on the nature of the data and the specific problem at hand. Here are five commonly used distance metrics:



```python
import numpy as np
import matplotlib.pyplot as plt

# Define two points
point1 = np.array([2, 3])  # Coordinates of point 1
point2 = np.array([5, 7])  # Coordinates of point 2

# Calculate distances
euclidean_distance = np.linalg.norm(point1 - point2)
manhattan_distance = np.sum(np.abs(point1 - point2))
cosine_similarity = np.dot(point1, point2) / (np.linalg.norm(point1) * np.linalg.norm(point2))
hamming_distance = np.count_nonzero(point1 != point2)
minkowski_distance = np.power(np.sum(np.power(np.abs(point1 - point2), 5)), 1/5)

# Print distances
print("Euclidean Distance:", euclidean_distance)
print("Manhattan Distance:", manhattan_distance)
print("Cosine Similarity:", cosine_similarity)
print("Hamming Distance:", hamming_distance)
print("Minkowski Distance:", minkowski_distance)

# Plot the points
plt.figure(figsize=(10, 8))
plt.scatter(point1[0], point1[1], color='blue', label='Point 1')
plt.scatter(point2[0], point2[1], color='red', label='Point 2')

# Mark the points
plt.text(point1[0], point1[1], '  (2, 3)', verticalalignment='bottom', horizontalalignment='right')
plt.text(point2[0], point2[1], '  (5, 7)', verticalalignment='bottom', horizontalalignment='right')

# Plot lines for Euclidean distance
plt.plot([point1[0], point2[0]], [point1[1], point2[1]], linestyle='--', color='green', label='Euclidean Distance')

# Plot lines for Manhattan distance
plt.plot([point1[0], point2[0]], [point1[1], point1[1]], linestyle='--', color='purple')
plt.plot([point2[0], point2[0]], [point1[1], point2[1]], linestyle='--', color='purple', label='Manhattan Distance')

# Annotate Cosine Similarity
plt.text((point1[0] + point2[0]) / 2, (point1[1] + point2[1]) / 2, 'Cosine Sim. = {:.2f}'.format(cosine_similarity), color='black')

# Annotate Hamming Distance
plt.text(point1[0] + 0.1, point2[1] - 0.1, 'Hamming Dist. = {}'.format(hamming_distance), color='black')

# Annotate Minkowski Distance
plt.text(point1[0] + 0.1, point1[1] - 0.1, 'Minkowski Dist. = {:.2f}'.format(minkowski_distance), color='black')

# Set plot attributes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Points and Distances')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.axis('equal')
plt.show()
```
