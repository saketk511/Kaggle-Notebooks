```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/travel-dataset-guide-to-indias-must-see-places/Top Indian Places to Visit.csv")
df.head()
```

```python
# 1. Bar Plot
sns.barplot(x='Significance', y='Number of google review in lakhs', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# 2. Scatter Plot

# Create scatter plot
sns.scatterplot(x='Number of google review in lakhs', y='Google review rating', hue='Significance', data=df)

# Position the legend outside the plot area on the right
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Adjust the margins to make space for the legend
plt.subplots_adjust(right=0.75)

# Show the plot
plt.show()
```

```python
# 3. Histogram
sns.histplot(df['Google review rating'], kde=True)
plt.show()
```

```python
# 4. Box Plot
sns.boxplot(x='Zone', y='time needed to visit in hrs', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Violin Plot

sns.violinplot(x='Zone', y='Entrance Fee in INR', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Heatmap

corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 7. Pair Plot

sns.pairplot(df)
plt.show()
```
